```python
from tika import parser

pdf_path = "./서비스업/[03204006]_035420.pdf"

# PDF 파일에서 텍스트를 추출
raw_pdf = parser.from_file(pdf_path) 
contents = raw_pdf['content'] 
contents = contents.strip()

print(contents)

file = open('Hello.txt','w')
file.write(raw_pdf['content'] )
file.close()
```
